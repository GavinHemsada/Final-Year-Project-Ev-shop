```python
import joblib
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType
import numpy as np
import os
```

```python
model = joblib.load("models/battery_health_model.pkl")
feature_cols = joblib.load("models/battery_health_features.pkl")
```

```python
initial_type = [
    ('float_input', FloatTensorType([None, len(feature_cols)]))
]
```

```python
onnx_model = convert_sklearn(
    model,
    initial_types=initial_type,
    target_opset=12
)
```

```python
os.makedirs('ONNXmodels', exist_ok=True)
```

```python
with open("ONNXmodels/battery_health_model.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```
